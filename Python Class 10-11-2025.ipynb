```python
class Student:
    def __init__(self, name, marks):
        
        self.name = name
        self.marks = marks

    def display(self):
        print(f"\nStudent Name: {self.name}")
        print(f"Marks in Python: {self.marks[0]}")
        print(f"Marks in Gen AI: {self.marks[1]}")
        print(f"Marks in Data Analytics: {self.marks[2]}")
```

```python
students = []

n = int(input("Enter number of students: "))

for i in range(n):
    print(f"\nEnter details for student {i+1}:")
    name = input("Enter name: ")
    python_marks = float(input("Enter marks in Python: "))
    gen_ai_marks = float(input("Enter marks in Gen AI: "))
    da_marks = float(input("Enter marks in Data Analytics: "))

    marks = [python_marks, gen_ai_marks, da_marks]

    students.append(Student(name, marks))
```

```python
print("\n===== STUDENT DETAILS =====")
for student in students:
    student.display()
```
