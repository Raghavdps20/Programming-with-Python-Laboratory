Q1) Generate all prime numbers less than 100.Print them in a spiral format (like a clockwise
square spiral) of minimal size that fits all primes.

```python
import math

def is_prime(n):
    """Check if a number is prime"""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

def generate_primes(limit):
    """Generate all prime numbers less than the given limit"""
    primes = []
    for num in range(2, limit):
        if is_prime(num):
            primes.append(num)
    return primes

def create_spiral(primes):
    """Create a spiral pattern with the given primes"""
    n = len(primes)
    
    # Find the smallest square grid that can fit all primes
    size = math.ceil(math.sqrt(n))
    
    # Initialize grid with empty strings
    grid = [['' for _ in range(size)] for _ in range(size)]
    
    # Directions: right, down, left, up
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    dir_idx = 0
    
    # Starting position
    row, col = size // 2, size // 2
    
    # If size is even, adjust starting position
    if size % 2 == 0:
        row -= 1
    
    prime_index = 0
    
    # Fill the spiral
    for step in range(1, size * 2):
        steps_in_direction = (step + 1) // 2
        
        for _ in range(steps_in_direction):
            if 0 <= row < size and 0 <= col < size and prime_index < n:
                grid[row][col] = str(primes[prime_index])
                prime_index += 1
            
            # Move to next position
            if prime_index < n:
                dr, dc = directions[dir_idx]
                row += dr
                col += dc
        
        # Change direction
        dir_idx = (dir_idx + 1) % 4
    
    return grid

def print_spiral(grid):
    """Print the spiral grid"""
    # Find the maximum width for formatting
    max_width = max(len(cell) for row in grid for cell in row if cell != '')
    
    print(f"Prime Numbers Spiral (n={len(grid)}x{len(grid)}):")
    print("=" * (len(grid) * (max_width + 3)))
    
    for row in grid:
        for cell in row:
            if cell == '':
                print(' ' * (max_width + 2), end=' ')
            else:
                print(f'{cell:>{max_width}}', end='  ')
        print()
    
    print("=" * (len(grid) * (max_width + 3)))



primes = generate_primes(100)
print(f"Prime numbers less than 100: {primes}")
print(f"Total primes: {len(primes)}")
print()

spiral_grid = create_spiral(primes)
print_spiral(spiral_grid)
```

Q2) You are a data analyst at a streaming platform, managing user interactions, analytics,
and daily operations. Use a for loop to create a text-based graph of hourly views. For
example, use * to represent every 5 views.

```python
# Sample hourly views data
hourly_views = [12, 25, 37, 8, 50, 63, 29, 10, 5, 42]

print("Hourly Views Graph ( * = 5 views )\n")

# Using a for loop to create the graph
for hour, views in enumerate(hourly_views, start=1):
    stars = "*" * (views // 5)   # one star for every 5 views
    print(f"Hour {hour:02d}: {stars} ({views} views)")
```

Q3) Each video watched generates revenue based on ad impressions: 
<br> First 10 views: 0.50 per view <br>
Next 20 views: 0.30 per view
<br> Remaining views: $0.10 per view
<br> You are given hourly views for the day. Calculate total daily revenue.
                               

```python
def calculate_daily_revenue(hourly_views):
    total_views = sum(hourly_views)
    revenue = 0.0

    # First 10 views
    if total_views <= 10:
        revenue = total_views * 0.50
    else:
        revenue += 10 * 0.50
        remaining_views = total_views - 10

        # Next 20 views
        if remaining_views <= 20:
            revenue += remaining_views * 0.30
        else:
            revenue += 20 * 0.30
            remaining_views -= 20

            # Remaining views
            revenue += remaining_views * 0.10

    return revenue

hourly_views = [5, 12, 8, 6, 10, 15, 7, 9, 11, 10, 5, 4, 8, 6, 7, 9, 10, 12, 6, 8, 7, 5, 9, 10]  # hourly views for 24 hours
total_revenue = calculate_daily_revenue(hourly_views)
print(f"Total daily revenue: ${total_revenue:.2f}")
```
