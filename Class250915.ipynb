```python
n = int(input("Enter the num : "))
```

```python
sum = 0
for i in range(1,n+1):
    print(f"{i}/{i+1}",end = " ")
    sum += i/(i+1)
    if i != n:
        print("+", end = " ")

print("=", sum)
```

```python
n = int(input("Enter the num : "))
```

```python
sum = 0
for i in range(1,n+1):
    print(f"{i}^{2}/{i}",end = " ")
    sum += (i**2)/i
    if i != n:
        print("+", end = " ")

print("=", sum)
```

```python
m1 = int(input("Enter number of rows of matrix 1: "))
n1 = int(input("Enter number of columns of matrix 1: "))
```

```python
m2 = int(input("Enter number of rows of matrix 2: "))
n2 = int(input("Enter number of columns of matrix 2: "))
```

```python
if(n1 == m2):
    print("Matrix multiplication can be performed")
else:
    print("Matric multiplication cannot be performed")
```

```python
a = [[1,4,6],
     [2,3,5],
     [7,2,5]]

b = [[3,4,6],
     [5,8,3],
     [4,9,4]]
```

```python
c = [[0 for _ in range(0,n2)] for _ in range(0,m1)]
```

```python
c
```

```python
for i in range(m1):
    for j in range(n2):
        for k in range(n1):
            c[i][j] += a[i][k] * b[k][j]
```

```python
for row in c:
    print(row)
```
