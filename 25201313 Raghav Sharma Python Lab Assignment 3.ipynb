WA function called Remainder that takes in paramenter x,y where y defaults to 4 and returns the remaider of x/y and ingores cases when x is negative or y is less than or eguql to zero

```python
def Remainder(x, y=4):
    if x < 0 or y <= 0:
        return None
    return x % y

a = int(input("Input the first number : "))
b = int(input("Input the second number : "))

result = Remainder(a,b)
print("Remainder = " , result)
```

Write a python function that print out the first n rows of Pascal triangle

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact * i
    return fact

n = int(input("Enter the value of n : "))
for i in range(n):
    for j in range(n-i-1):
        print(" ", end="")

    for j in range(i+1):
        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")

    print()
```

WA Function sumPdivisor() that finds the sum of proper divisors of a number

```python
def sumPdivisor(n):
    sum = 0
    for i in range(1, n):
        
        if n % i == 0:
            sum += i
    
    return sum

n = int(input("Enter a number : "))

result = sumPdivisor(n)
print(result)
```

WA menu driven program using functions to perform calculation function add, sub, mul, div of two integers

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b

def menu():
    while True:
        print("\n=== Calculator Menu ===")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        
        choice = input("Enter your choice (1-5): ")

        if choice == '5':
            break

        num1 = int(input("Enter first integer: "))
        num2 = int(input("Enter second integer: "))


        if choice == '1':
            print("Result:", add(num1, num2))
        elif choice == '2':
            print("Result:", sub(num1, num2))
        elif choice == '3':
            print("Result:", mul(num1, num2))
        elif choice == '4':
            print("Result:", div(num1, num2))
        else:
            print("Invalid choice!")

menu()
```

WAF using C(n,r) to calculate the conpound interest for the given principle, rate of interest and number of years and return the value

```python
def comb(i,j):
    return factorial(i)//(factorial(j)*factorial(i-j))

def compound_interest(principal, rate, years):
    
    r = rate / 100
    total = 0
    
    for k in range(years + 1):
        total += comb(years, k) * (r ** k)
    
    return principal * total


P = 1000
R = 5
N = 3
print("Final Amount:", compound_interest(P, R, N))
```
