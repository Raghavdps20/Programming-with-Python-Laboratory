```python
import itertools

d = {'1': ['a', 'b'], '2': ['c', 'd']}

for combo in itertools.product(*[d[k] for k in sorted(d.keys())]):
    print(''.join(combo))
```

```python
data = {'a': 5, 'b': 14, 'c': 32, 'd': 35, 'e': 24, 
        'f': 100, 'g': 57, 'h': 8, 'i': 100}

def max_n_values(dictionary, n):
    sorted_items = sorted(dictionary.items(), key=lambda x: x[1], reverse=True)
    top_n = sorted_items[:n]
    return [key for key, value in top_n]

print("Original Dictionary:")
print(data)

print("\n1 maximum value(s) in the said dictionary:")
print(max_n_values(data, 1))

print("\n2 maximum value(s) in the said dictionary:")
print(max_n_values(data, 2))

print("\n5 maximum value(s) in the said dictionary:")
print(max_n_values(data, 5))
```

```python
def evaluate_expression(expr, variables):

    if "const" in expr:
        return expr["const"]
    
    if "var" in expr:
        var_name = expr["var"] 
        if var_name not in variables:
            raise ValueError(f"Variable '{var_name}' not defined")
        return variables[var_name]
    

    if "op" in expr:
        op = expr["op"] 
        left = evaluate_expression(expr["left"], variables)  
        right = evaluate_expression(expr["right"], variables) 
        
    
        if op == "+":
            return left + right 
        elif op == "-":
            return left - right  
        elif op == "*":
            return left * right  
        elif op == "/":
            if right == 0:
                raise ZeroDivisionError("Division by zero") 
            return left / right 
        else:
            raise ValueError(f"Unknown operator: {op}") 
    
    raise ValueError("Invalid expression format")


# x + y * 2
expr1 = {
    'op': '+', 
    'left': {'var': 'x'}, 
    'right': {'op': '*', 'left': {'var': 'y'}, 'right': {'const': 2}}  
}

# (10 / a) - (b * c)
expr2 = {
    'op': '-',  
    'left': {'op': '/', 'left': {'const': 10}, 'right': {'var': 'a'}},  
    'right': {'op': '*', 'left': {'var': 'b'}, 'right': {'var': 'c'}} 
}

variables1 = {'x': 5, 'y': 3}

variables2 = {'a': 2, 'b': 3, 'c': 4}


result1 = evaluate_expression(expr1, variables1)

result2 = evaluate_expression(expr2, variables2)

print(f"Expression 1 result: {result1}")  
print(f"Expression 2 result: {result2}")
```
