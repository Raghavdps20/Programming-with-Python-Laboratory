2..Write a program to add ‘ing’ at the end of a given string. If the given string already endswith ‘ing’ and then add ‘ly’ instead. If the string length of the given string is less than 3, leave it unchanged.

```python
word = input("Enter a string")
```

```python
print(count)
```
