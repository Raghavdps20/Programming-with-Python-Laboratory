```python
# WAP that encrypts a message by adding a key value to every character
```

```python
message = input("Enter the message: ")
key = int(input("Enter the key value (integer): "))

encrypted_message = ""

for char in message:
    encrypted_message += chr(ord(char) + key)

print("Encrypted message:", encrypted_message)
```

```python
# WAP that finds whether a given character is present in a string or not, in case it is present you have to print the text, do not use a built in find function
```

```python
text = input("Enter a string: ")
ch = input("Enter a character to search for: ")

found = False  # flag variable

for c in text:
    if c == ch:
        found = True
        break

if found:
    print(f"The character '{ch}' is present in the text: {text}")
else:
    print(f"The character '{ch}' is NOT present in the text.")
```

```python
# WAP that counts the occurences of the character in that string, do not use the built in count function
```

```python
text = input("Enter a string: ")
ch = input("Enter a character to count: ")

count = 0

for c in text:
    if c == ch:
        count += 1

print(f"The character '{ch}' occurs {count} time(s) in the string.")
```

```python
# WA funtion to get a string made up of four copies of the last 2 characters of a specified string. (length must be atleast 2)
```

```python
def last_two_four_times(s):
    if len(s) < 2:
        return "String length must be at least 2."
    last_two = s[-2:]        
    return last_two * 4       

text = input("Enter a string (at least 2 characters): ")
result = last_two_four_times(text)
print("Result:", result)
```

```python
# WA function to reverse a string if its length is multiple of 4
```

```python
def reverse_if_multiple_of_4(s):
    if len(s) % 4 == 0:
        return s[::-1]  
    else:
        return s      

text = input("Enter a string: ")
result = reverse_if_multiple_of_4(text)
print("Result:", result)
```

```python
# WAP that uses a RE to match strings which starts with a sequence of digits (at least one digit) followed by a blank and after this some arbitrary characters
```

```python
import re

pattern = r'^\d+\s.*'

text = input("Enter a string: ")

if re.match(pattern, text):
    print("Match found!")
else:
    print("No match.")
```

```python
# WAP to print only first 2 character of every word. do it using both RE and normal method
```

```python
text = input("Enter a sentence: ")

words = text.split()

for w in words:
    print(w[:2], end=' ')
```

```python
import re

text = input("Enter a sentence: ")

result = re.findall(r'\b\w{1,2}', text)

print("First 2 characters of each word:")
for word in result:
    print(word, end=' ')
```
