```python
# Q1. Write a program that has classes such as Student, Course and Department. Enroll a student in a course of a particular department.
```

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.courses = []

    def enroll(self, course):
        self.courses.append(course)
        print(f"{self.name} has been enrolled in {course.course_name}.")


class Course:
    def __init__(self, course_name, course_code):
        self.course_name = course_name
        self.course_code = course_code
        self.students = []

    def add_student(self, student):
        self.students.append(student)
        print(f"Student {student.name} added to course {self.course_name}.")


class Department:
    def __init__(self, dept_name):
        self.dept_name = dept_name
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)
        print(f"Course {course.course_name} added to {self.dept_name} department.")

    def enroll_student_in_course(self, student, course_code):
        for course in self.courses:
            if course.course_code == course_code:
                student.enroll(course)
                course.add_student(student)
                return
        print("Course not found in this department.")



cs_dept = Department("Computer Science")

course1 = Course("Data Structures", "CS101")
course2 = Course("Algorithms", "CS102")

cs_dept.add_course(course1)
cs_dept.add_course(course2)

student1 = Student("Alice", "S001")

cs_dept.enroll_student_in_course(student1, "CS101")
```

```python
# Q2. Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.
```

```python
class Bill:
    def __init__(self, amount):
        self.amount = amount

    def pay(self):
        print("Payment method not specified.")


# Derived class: Cash Payment
class CashPayment(Bill):
    def __init__(self, amount):
        super().__init__(amount)

    def pay(self):
        print(f"Bill of {self.amount} has been paid in CASH.")


# Derived class: Cheque Payment
class ChequePayment(Bill):
    def __init__(self, amount, cheque_number):
        super().__init__(amount)
        self.cheque_number = cheque_number

    def pay(self):
        print(f"Bill of {self.amount} has been paid by CHEQUE.")
        print(f"Cheque Number: {self.cheque_number}")


cash_bill = CashPayment(1500)
cash_bill.pay()

cheque_bill = ChequePayment(2500, "CHQ12345")
cheque_bill.pay()
```

```python
# Q3.Write a program to overload the - =operator to subtract two Distance Objects.
```

```python
class Distance:
    def __init__(self, meters=0, centimeters=0):
        self.meters = meters
        self.centimeters = centimeters
        self.__normalize()

    def __normalize(self):
        if self.centimeters >= 100:
            self.meters += self.centimeters // 100
            self.centimeters = self.centimeters % 100

        if self.centimeters < 0:
            borrow = (-self.centimeters + 99) // 100
            self.meters -= borrow
            self.centimeters += borrow * 100

    def __isub__(self, other):
        self.meters -= other.meters
        self.centimeters -= other.centimeters
        self.__normalize()
        return self

    def __str__(self):
        return f"{self.meters}m {self.centimeters}cm"


d1 = Distance(5, 80)   # 5m 80cm
d2 = Distance(2, 95)   # 2m 95cm

print("Before subtraction:")
print("d1 =", d1)
print("d2 =", d2)

d1 -= d2 

print("\nAfter d1 -= d2:")
print("d1 =", d1)
```

```python
# Q4. Write a program with two classes for calculating the distance. On has distance specified in meters and other has distance in kilometres. These functions take argument of class Distance that converts the distance into kilometres and meters.
```

```python
class DistanceMeters:
    def __init__(self, meters=0):
        self.meters = meters

    def to_kilometers(self):
        """Convert meters to kilometers."""
        return self.meters / 1000

    def show(self):
        print(f"Distance: {self.meters} meters")


class DistanceKilometers:
    def __init__(self, km=0):
        self.km = km

    def to_meters(self):
        """Convert kilometers to meters."""
        return self.km * 1000

    def show(self):
        print(f"Distance: {self.km} kilometers")


d1 = DistanceMeters(500)
d1.show()
print("In kilometers:", d1.to_kilometers())

d2 = DistanceKilometers(3)
d2.show()
print("In meters:", d2.to_meters())
```
