```python
# WAP that creates a list of numbers from 1-20 that are either divisible by 2 or divisible by 4
```

```python
numbers = [n for n in range(1, 21) if n % 2 == 0 or n % 4 == 0]
print(numbers)
```

```python
# WAP to list a squares of number from 1-10. Then use for ... in constuct to sum elements in list generated.
```

```python
squares = [n**2 for n in range(1, 11)]

total = 0

for num in squares:
    total += num

print("List of squares:", squares)
print("Sum of squares:", total)
```

```python
# WAP that defines a list of countries that are a member of BRICS. Check whether a country is a memeber of BRICS or not.
```

```python
brics = ["Brazil", "Russia", "India", "China", "South Africa"]

country = input("Enter the name of a country: ")

if country.capitalize() in brics:
    print(f"{country} is a member of BRICS.")
else:
    print(f"{country} is not a member of BRICS.")
```

```python
''' Simulate and analyze traffic movement on a single road with several lanes using lists
Generate random vehicles
Simulate entry and exit 
Measure congestion, flow, averge wait time.'''
```

```python
import random 

num_lanes = 4
threshold = 5

lanes = []
for i in range(num_lanes):
    lanes.append([])

for lane in lanes:
    num_vehicles = random.randint(0, 8)
    
    for v in range(num_vehicles):
        vehicle_id = "V" + str(random.randint(100, 999))
        lane.append(vehicle_id)

total_vehicles = 0
for lane in lanes:
    total_vehicles += len(lane)

congested_lanes = 0
for lane in lanes:
    if len(lane) > threshold:
        congested_lanes += 1

average_vehicles = total_vehicles / num_lanes

for i, lane in enumerate(lanes, start=1):
    if len(lane) > threshold:
        status = "Congested"
    else:
        status = "Clear"
    print(f"Lane {i}: {len(lane)} vehicles -> {status}")

# Print overall results
print("\nTotal vehicles:", total_vehicles)
print("Congested lanes:", congested_lanes, "/", num_lanes)
print("Average vehicles per lane:", round(average_vehicles, 2))
```
