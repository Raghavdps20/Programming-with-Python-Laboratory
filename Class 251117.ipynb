```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def display(self):
        print("Name :" , self.name, ", Age :", self.age)
```

```python
class Teacher(Person):
    def __init__(self,name,age,exp,res):
        Person.__init__(self,name,age)
        self.exp = exp
        self.res = res

    def displayData(self):
        Person.display(self)
        print("Exp :", self.exp, ", Research : " , self.res)
```

```python
class Student(Person):
    def __init__(self, name, age, course, marks):
        Person.__init__(self,name,age)
        self.course = course
        self.marks = marks

    def displayData(self):
        Person.display(self)
        print("Course : ", self.course, ", Marks" , self.marks)
```

```python
P = Person("Max", 20)
P.display()
```

```python
T = Teacher("Stark", 40, 10, "AI CS")
T.displayData()
```

```python
S = Student("Rahul", 22, "AI", 99)
S.displayData()
```
