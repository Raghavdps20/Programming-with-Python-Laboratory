```python
# WAP to calculate area of a trinangle using heron's formula
a = float(input("Enter a = "))
b = float(input("Enter b = "))
c = float(input("Enter c = "))
s = (a+b+c)/2
area = ((s)*(s-a)*(s-b)*(s-c))**0.5
print("Area = ", area)
```

```python
# Write a program to perform addition, subtraction, division, multiplication of two floating point numbers
a = float(input("Enter num1 : "))
b = float(input("Enter num2: "))
op = ['+', '-', '*', '/', '//', 'exit']
i=0

while(True):
    # c = input("Enter operation +, -, *, / , //:")
    c = op[i]
    i += 1
    match c:
        case '+':
            print("a + b = " + str(a+b))
        case '-':
            print("a - b = " + str(a-b))
        case '*':
            print("a * b = " + str(a*b))
        case '/':
            print("a / b = " + str(a/b))
        case '//':
            print("a // b = " + str(a//b))
        case 'exit':
            break
```

```python
# Write a program to calculate bill amount for a item given its quantity sold value, discount and tax

qs = 10
ppi = 5
amt = qs*ppi
discount = 0.10
discount_total = discount * amt
tax = 0.18
total_amt = (amt - discount_total)*(1+tax)

print(format("BILL", '*>16') + format("", '*>12'))
print("Qauntity Sold:\t\t ", qs)
print("Price per item:\t\t  ", ppi)

print(format("", '->28'))
print("Amount: \t\t ", amt)
print("Discount:\t      ", str(discount*100)+'%')
print(format("", '->28'))
print("Discount total:\t\t", discount_total)
print("Tax:\t\t      ", str(tax*100)+'%')
print(format("", '->28'))
print("Total amt to be paid: ", "{:.2f}".format(total_amt))
```

```python
# Write a program to calculate student's result based on two examinations, 1 Sport event and 3 activities conducted. 
# Weightage of activities 30%, sports 20% & examination = 50%
marks_exam_1 = float(input("Enter marks for Exam 1: "))
marks_exam_2 = float(input("Enter marks for Exam 2: "))
marks_sports = float(input("Enter marks for Sports: "))
marks_activities_1 = float(input("Enter marks for Activity 1: "))
marks_activities_2 = float(input("Enter marks for Activity 2: "))
marks_activities_3 = float(input("Enter marks for Activity 3: "))
final = (marks_exam_1+marks_exam_2)*0.3/2 + (marks_sports) *0.2 + (marks_activities_1+marks_activities_2+marks_activities_3)*0.5/3 
print("Final Result : ", final)
```
